```python
import os
import random
import copy
import pandas as pd
import numpy as np

workingDir = 'G:\My Drive\Term 4.2\IS706 - Software Mining and Analysis\\Programming Assignment 2\\repo\\IS706-Assignment2\\resources\\documents'

# workingDir='./resources/documents'

os.chdir(workingDir)

documentNames = [doc for doc in os.listdir() if '.txt' in doc] # to remove the desktop.ini
```

<h3>Create Incidence Dictionary

```python
docIncidenceDict = {documentName:0 for documentName in documentNames if '.txt' in documentName}

docIncidenceDict
```

<h3>Populate Incidence Dictionary for Every Term

```python
allDocTerms = {}

for documentName in documentNames:
    with open(documentName, 'r') as f:
        print('Processing', documentName)
        for line in f:
            for char in ['~','#','?','\\n','`','*','!']:
                line = line.replace(char, "")
            withoutCurly1 = line.replace('{', '')
            withoutCurly2 = withoutCurly1.replace('}', '')
            withoutEqual = withoutCurly2.replace('=', ' ')
            withoutBar = withoutEqual.replace('|', '')
            withoutAt = withoutBar.replace('@','')
            withoutPlus = withoutAt.replace('+',' ')
            withoutMinus = withoutPlus.replace('-',' ')
            withoutParan = withoutMinus.replace('(','').replace(')','')
            withoutSigns = withoutParan.replace('<','').replace('>','')
            withoutSymbols = withoutSigns.replace('_',' ').replace('/',' ').replace(':', ' ').replace('"', '').replace('\'','').replace('.',' ').replace(',','')
            terms = withoutSymbols.split()

            for term in terms:
                if not term.isnumeric() and term != '':
                    term = term.lower()
                    if term not in allDocTerms:
                        allDocTerms[term] = copy.deepcopy(docIncidenceDict)

                    allDocTerms[term][documentName] = allDocTerms[term][documentName] + 1
```

<h3>Term Frequency Function

```python
def termFrequency(docName):
    tfDict = {}

    for term in allDocTerms:
        term = term.lower()
#         if allTerms[term][docName] != 0:
        tfDict[term] = allDocTerms[term][docName]
    
    return tfDict    
```

```python
termFrequency('document32.txt')
```

<h3>Inverse Document Frequency Function

```python
n = len(documentNames)
idf={k:(n/(sum([1 for e in v.values() if e != 0]))) for k,v in  allDocTerms.items()} 
   
idf['crypto']
```

<h3>Creating Vector Space Representation for all Documents</h3>

```python
zeroes = [0.0]*len(idf)

temp={'words_idf':idf.values()}
for d in documentNames:
    temp[d.strip('.txt')] = zeroes

df = pd.DataFrame(temp, index=idf.keys())

df
```

<h3>Calculate TF-IDF embedding for all Document Terms

```python
for d in documentNames:
    temp = termFrequency(d)
    d = d.strip('.txt')
    for word, tf in temp.items():
        df[d][word] = df['words_idf'][word] * float(tf)
```

<h3>Combine the 2 sets of terms to have the same number of dimensions in the same vector space

```python
workingDir='../combined'

os.chdir(workingDir)
fileNames = [f for f in os.listdir() if '.txt' in f]

combinedIncidenceDict = {f:0 for f in fileNames}
print(len(combinedIncidenceDict), combinedIncidenceDict)
```

<h3>Populate Incidence Dictionary for both Document and Query Terms

```python
allCombinedTerms = {}

for fileName in fileNames:
    with open(fileName, 'r') as f:
        print('Processing', fileName)
        for line in f:
            for char in ['~','#','?','\\n','`','*','!']:
                line = line.replace(char, "")
            withoutCurly1 = line.replace('{', '')
            withoutCurly2 = withoutCurly1.replace('}', '')
            withoutEqual = withoutCurly2.replace('=', ' ')
            withoutBar = withoutEqual.replace('|', '')
            withoutAt = withoutBar.replace('@','')
            withoutPlus = withoutAt.replace('+',' ')
            withoutMinus = withoutPlus.replace('-',' ')
            withoutParan = withoutMinus.replace('(','').replace(')','')
            withoutSigns = withoutParan.replace('<','').replace('>','')
            withoutSymbols = withoutSigns.replace('_',' ').replace('/',' ').replace(':', ' ').replace('"', '').replace('\'','').replace('.',' ').replace(',','')
            terms = withoutSymbols.split()

            for term in terms:
                if not term.isnumeric() and term != '':
                    term = term.lower()
                    if term not in allCombinedTerms:
                        allCombinedTerms[term] = copy.deepcopy(combinedIncidenceDict)

                    allCombinedTerms[term][fileName] = allCombinedTerms[term][fileName] + 1
```

```python
def combinedTermFrequency(name):
    tfDict = {}

    for term in allCombinedTerms:
        term = term.lower()
        tfDict[term] = allCombinedTerms[term][name]
    
    return tfDict    
```

```python
n = len(fileNames)
combinedIdf={k:(n/(sum([1 for e in v.values() if e != 0]))) for k,v in  allCombinedTerms.items()} 
combinedIdf
```

<h3>Calculating the TF-IDF Embedding across all the Documents and Queries </h3>

```python
zeroes = [0.0]*len(combinedIdf)
temp={'words_combinedIdf':combinedIdf.values()}
for d in fileNames:
    temp[d.strip('.txt')] = zeroes

combined_df = pd.DataFrame(temp, index=combinedIdf.keys())

combined_df
```

```python
for f in fileNames:
    temp = combinedTermFrequency(f)
    f = f.strip('.txt')
    for word, tf in temp.items():
        combined_df[f][word] = combined_df['words_combinedIdf'][word] * float(tf)

combined_df
```

<h3>Calculating the cosine similarity between each query and document </h3>

```python
from scipy import spatial

def cos_sim(v1,v2):
    return 1 - spatial.distance.cosine(v1,v2)

total_similarities = {}
for query in queryNames:
    query=query.strip('.txt')
    cur = {}
    for doc in documentNames:
        doc=doc.strip('.txt')
        cur[doc] = cos_sim(combined_df[query], combined_df[doc])
    total_similarities[query] = dict(sorted(cur.items(), key=lambda x: x[1], reverse=True))

total_similarities
```
