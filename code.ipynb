```python
import os
import random
import copy
import pandas as pd
import numpy as np

# workingDir = 'G:\My Drive\Term 4.2\IS706 - Software Mining and Analysis\\Programming Assignment 2\\repo\\IS706-Assignment2\\resources\\documents'

workingDir='./resources/documents'

os.chdir(workingDir)
documentNames = os.listdir()
documentNames = documentNames[1:] # to remove the desktop.ini


workingDir='../queries'

os.chdir(workingDir)
queryNames = os.listdir()
```

<h3>Create Incidence Dictionary

```python
# incidencyDict = {}

# for documentName in documentNames:
#     incidencyDict[documentName] = 0
incidencyDict = {documentName:0 for documentName in documentNames if '.txt' in documentName}

incidencyDict
```

<h3>Populate Incidence Dictionary for Every Term

```python
allTerms = {}

for documentName in documentNames:
    with open(documentName, 'r') as f:
        print('Processing', documentName)
        for line in f:
            withoutCurly1 = line.replace('{', '')
            withoutCurly2 = withoutCurly1.replace('}', '')
            withoutEqual = withoutCurly2.replace('=', ' ')
            withoutBar = withoutEqual.replace('|', '')
            withoutAt = withoutBar.replace('@','')
            withoutPlus = withoutAt.replace('+',' ')
            withoutMinus = withoutPlus.replace('-',' ')
            withoutParan = withoutMinus.replace('(','').replace(')','')
            withoutSigns = withoutParan.replace('<','').replace('>','')
            withoutSymbols = withoutSigns.replace('_',' ').replace('/',' ').replace(':', ' ').replace('"', '').replace('\'','').replace('.',' ')
            terms = withoutSymbols.split()

            for term in terms:
                if not term.isnumeric() and term != '':
                    term = term.lower()
                    if term not in allTerms:
                        allTerms[term] = copy.deepcopy(incidencyDict)

                    allTerms[term][documentName] = allTerms[term][documentName] + 1
```

<h3>Term Frequency Function

```python
def termFrequency(docName):
    tfDict = {}

    for term in allTerms:
        term = term.lower()
#         if allTerms[term][docName] != 0:
        tfDict[term] = allTerms[term][docName]
    
    return tfDict    
```

```python
termFrequency('document32.txt')
```

<h3>Inverse Document Frequency Function

```python
n = len(documentNames)
idf={k:(n/(sum([1 for e in v.values() if e != 0]))) for k,v in  allTerms.items()} 
   
idf['crypto']
```

```python
def inverseDocumentFrequency(docName):
    idfDict = {}

    for term in allTerms:
        term = term.lower()
        if allTerms[term][docName] != 0:
            idfDict[term] = 0

    for term in idfDict:
        idfDict[term] = len(documentNames)/sum(1 for v in allTerms[term].values() if v != 0)
    
    
    return dict(sorted(idfDict.items(), key=lambda item: item[1], reverse=True)) 
```
